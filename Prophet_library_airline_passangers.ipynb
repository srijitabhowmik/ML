```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
airline = pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
airline
```

```python
airline.dropna(inplace=True)
airline
```

```python
airline.head()
```

Simple Moving Average

```python
airline['6-months-SMA']=airline['Thousands of Passengers'].rolling(window=6).mean()
airline['12-months-SMA']=airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline.head(17)
```

```python
airline.plot();
```

Exponentially Weighted oving Average [EWMA]

```python
airline['EWMA12'] = airline['Thousands of Passengers'].ewm(span=12,adjust=False).mean()
```

```python
airline[['Thousands of Passengers','EWMA12']].plot();
```

```python
airline[['Thousands of Passengers','EWMA12','12-months-SMA']].plot(figsize=(12,8)).autoscale(axis='x',tight=True);
```
